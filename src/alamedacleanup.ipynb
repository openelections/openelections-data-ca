```python
import pandas as pd
import numpy as np
```

```python
#load data
df = pd.read_csv('alamedadata.csv')
```

```python
df
```

```python
#drop unnecessary columns
df = df.drop(['REPORTING_FLAG','CANDIDATE_ID','PRECINCT_NAME', 'CONTEST_ID','CONTEST_SEQ', 'VOTES_ALLOWED'], axis=1)
```

```python
#Replace party code with full name (could be done with dictionary but did this after I learned how to do that) 

df = df.replace({'PARTY_CODE': 1}, 'Democrat')
df = df.replace({'PARTY_CODE': 2}, 'Republican')
df = df.replace({'PARTY_CODE': 3}, 'Independent')
df = df.replace({'PARTY_CODE': 4}, 'Green')
df = df.replace({'PARTY_CODE': 5}, 'Libertarian')
df = df.replace({'PARTY_CODE': 6}, 'Peace and Freedom')

# df['PARTY_CODE'] = {'1': 'Democrat', '2':}
# partymap['1']
# #dictionary, lambda apply
```

```python
#renaming positions and removing political party

df = df.replace({'CONTEST_FULL_NAME': 'AI - Presidential Preference'}, 'President')
df = df.replace({'CONTEST_FULL_NAME': 'DEM - Presidential Preference'}, 'President')
df = df.replace({'CONTEST_FULL_NAME': 'REP - Presidential Preference'}, 'President')
df = df.replace({'CONTEST_FULL_NAME': 'LIB - Presidential Preference'}, 'President')
df = df.replace({'CONTEST_FULL_NAME': 'GRN - Presidential Preference'}, 'President')
df = df.replace({'CONTEST_FULL_NAME': 'PF - Presidential Preference'}, 'President')
df = df.replace({'CONTEST_FULL_NAME': 'GRN - County Council  At Large'}, 'County Council At Large')
df = df.replace({'CONTEST_FULL_NAME': 'DEM - County Central Committee  15th Assembly District'}, 'County Central Committee 15th Assembly District')
df = df.replace({'CONTEST_FULL_NAME': 'PF - County Central Committee  5th Supervisorial District'}, 'County Central Committee  5th Supervisorial District')
df = df.replace({'CONTEST_FULL_NAME': 'DEM - County Central Committee  18th Assembly District'}, 'County Central Committee  18th Assembly District')
df = df.replace({'CONTEST_FULL_NAME': 'DEM - County Central Committee  20th Assembly District'}, 'County Central Committee  20th Assembly District')
df = df.replace({'CONTEST_FULL_NAME': 'REP - County Central Committee  20th Assembly District'}, 'County Central Committee  20th Assembly District')
df = df.replace({'CONTEST_FULL_NAME': 'DEM - County Central Committee  16th Assembly District'}, 'County Central Committee  16th Assembly District')
df = df.replace({'CONTEST_FULL_NAME': 'REP - County Central Committee  16th Assembly District'}, 'County Central Committee  16th Assembly District')
df = df.replace({'CONTEST_FULL_NAME': 'DEM - County Central Committee  25th Assembly District'}, 'County Central Committee  25th Assembly District')
```

```python
#rename columns again
df = df.rename(columns={"CONTEST_FULL_NAME": "office", "VOTE_COUNT": "votes","CANDIDATE_FULL_NAME":"candidate", "PARTY_CODE": "party", "PRECINCT_ID":"precinct", "BALLOT_CAST":"ballots_cast","REGISTER_VOTERS":"registered_voters"})
```

```python
#filtering the office into positions with district in name (not included in the contain func)
dfccc = df[df['office'].str.contains("County Central Committee")]
dfusr=  df[df['office'].str.contains("U.S. Representative")]
dfss= df[df['office'].str.contains("State Senator")]
dfsa= df[df['office'].str.contains("State Assembly")]
dfsup= df[df['office'].str.contains("Supervisor ")]
dfboe= df[df['office'].str.contains("Board of Education")]
dfusdd= df[df['office'].str.contains("Union Sanitary District Director")]

dfsup
```

```python
#line one: splits ['office'] into district
#line two: adds the latter end (that includes the district) to it's own column
#note: it gives a weird error, but it still does what I need it to do so I've continued with this message 


cccsplit = dfccc["office"].str.split(" ", n = 3, expand = True) 
dfccc['district']= cccsplit[3] #dfccc['districts'] contains all districts for county central committee

usrsplit = dfusr["office"].str.split(" ", n=2, expand = True) 
dfusr["district"]= usrsplit[2] #dfusr'districts'] contains all districts for us representatives

sssplit = dfss["office"].str.split(" ", n=2, expand = True) 
dfss["district"]= sssplit[2] #dfss['districts'] contains all districts for state senate

sasplit = dfsa["office"].str.split(" ", n=2, expand = True) 
dfsa["district"]= sasplit[2] #dfsa['districts'] contains all districts for state assembly

supsplit = dfsup["office"].str.split(" ", n=1, expand = True) 
dfsup["district"]= supsplit[1] #dfsup['districts'] contains all districts for supervisor

boesplit = dfboe["office"].str.split(" ", n=3, expand = True) 
dfboe["district"]= boesplit[3] #dfboe['districts'] contains all districts for board of education

usddsplit = dfusdd["office"].str.split(" ", n=4, expand = True) 
dfusdd["district"]= usddsplit[4] #dfboe['districts'] contains all districts for board of education
```

```python
#dropping unneeded column names from new dataframe

dfccc = dfccc.drop(['precinct', 'office', 'candidate', 'party', 'votes'], axis=1)
dfusr = dfusr.drop(['precinct', 'office', 'candidate', 'party', 'votes'], axis=1)
dfss = dfss.drop(['precinct', 'office', 'candidate', 'party', 'votes'], axis=1)
dfsa = dfsa.drop(['precinct', 'office', 'candidate', 'party', 'votes'], axis=1)
dfsup = dfsup.drop(['precinct', 'office', 'candidate', 'party', 'votes'], axis=1)
dfboe = dfboe.drop(['precinct', 'office', 'candidate', 'party', 'votes'], axis=1)
dfusdd = dfusdd.drop(['precinct', 'office', 'candidate', 'party', 'votes'], axis=1)
```

```python
df2 = df.join(dfccc[['district']])
df2['district'].update(dfusr['district'])
df2['district'].update(dfss['district'])
df2['district'].update(dfsa['district'])
df2['district'].update(dfsup['district'])
df2['district'].update(dfboe['district'])
df2['district'].update(dfusdd['district'])
df2.loc[96]
```
